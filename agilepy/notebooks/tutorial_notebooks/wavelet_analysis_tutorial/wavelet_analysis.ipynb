# Wevelet Analysis notebook

## Prerequisities: in order to run this notebook you need ctsmap generated by AGAnalysis class

```python
from agilepy.api.AGAnalysisWavelet import AGAnalysisWavelet
```

```python
# Interactive plots
%matplotlib widget
```

## Creating a configuration file

```python
confFile = "./agilepy_wavelet.yaml"
```

```python
AGAnalysisWavelet.getConfiguration(
        confFilePath=confFile,
        userName="user",
        outputDir="$HOME/agilepy_analysis",
        verboselvl=2,
        ctsmap="./../../../testing/unittesting/api/data/analysis_product_EMIN00100_EMAX10000_01.cts.gz",
        scaletype="dyadic",
        scalemin=0.031250,
        scalemax=1024,
        scalenum=46,
        methistsize=1000,
        cclsizemin=-1,
        cclsizemax=-1,
        cclradmin=-1,
        cclradmax=-1,
        cclscalemin=-1,
        cclscalemax=-1
        )
```

## Loading configuration file and class instance

```python
ag = AGAnalysisWavelet(confFile)
```

## Wavelet analysis

The analysis produces three files:
- .twf file:
- .met file:
- .ccl file:

```python
f1, f2, f3 = ag.waveletAnalysis()
```

## Display
